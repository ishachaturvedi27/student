## Practice Problem 3 and your final mission:

// A slightly more complex expression


// Step 1: Create three variables and give them values

<strong> let x = 8;
let y = 3;
let z = 2; </strong>

// Step 2: Add x and y
// This will calculate (x + y)

<strong> let sumXY = x + y; // 8 + 3 = 11 </strong>

// Step 3: Multiply the result by z
// This will calculate (x + y) * z

<strong> let multiplied = sumXY * z; // 11 * 2 = 22 </strong>

// Step 4: Find the remainder when that result is divided by y
// This will calculate ((x + y) * z) % y

<strong> let finalResult = multiplied % y; // 22 % 3 = 1 </strong>

// Step 5: Print all the steps and the final result

<strong> console.log("x:", x);

console.log("y:", y);

console.log("z:", z);

console.log("x + y =", sumXY);

console.log("(x + y) * z =", multiplied);

console.log("Final result: ((x + y) * z) % y =", finalResult); </strong>
